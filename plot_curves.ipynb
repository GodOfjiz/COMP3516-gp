## Per-model curve (training loss on left axis, validation accuracy on right)

```python
def plot_run(tag, title, out_path):
    rows = RUNS[tag]
    if not rows:
        return
    eps  = [r[0] for r in rows]
    loss = [r[1] for r in rows]
    acc  = [r[2] for r in rows]

    fig, ax1 = plt.subplots(figsize=(6.0, 3.2))
    c1, c2 = "tab:blue", "tab:red"

    ax1.plot(eps, loss, color=c1, linewidth=1.4, label="Train loss")
    ax1.set_xlabel("Epoch")
    ax1.set_ylabel("Training loss", color=c1)
    ax1.tick_params(axis="y", labelcolor=c1)
    ax1.grid(True, linestyle="--", alpha=0.4)

    ax2 = ax1.twinx()
    ax2.plot(eps, acc, color=c2, linewidth=1.4, label="Val accuracy")
    ax2.set_ylabel("Validation accuracy", color=c2)
    ax2.tick_params(axis="y", labelcolor=c2)
    ax2.set_ylim(0.0, 1.0)

    best_i = max(range(len(acc)), key=lambda i: acc[i])
    ax2.scatter([eps[best_i]], [acc[best_i]], color="black", zorder=5, s=30,
                label=f"best={acc[best_i]:.3f} @ ep {eps[best_i]}")
    ax2.legend(loc="lower right", fontsize=8)

    plt.title(title)
    fig.tight_layout()
    plt.savefig(out_path, dpi=150, bbox_inches="tight")
    plt.show()
    print("Wrote", out_path)

plot_run("IRA",   "IRA LSTM: training loss and validation accuracy",         OUT_DIR / "curve_ira.png")
plot_run("wifi",  "WiFi LSTM: training loss and validation accuracy",        OUT_DIR / "curve_wifi.png")
plot_run("imu",   "IMU LSTM: training loss and validation accuracy",         OUT_DIR / "curve_imu.png")
plot_run("multi", "Multimodal LSTM: training loss and validation accuracy",  OUT_DIR / "curve_multi.png")
```

## Combined validation-accuracy comparison across models

```python
fig, ax = plt.subplots(figsize=(6.2, 3.4))
styles = [("IRA", "IRA", "-"), ("wifi", "WiFi", "--"),
          ("imu", "IMU", "-."), ("multi", "Multimodal", ":")]
for tag, label, style in styles:
    rows = RUNS[tag]
    if not rows:
        continue
    eps = [r[0] for r in rows]
    acc = [r[2] for r in rows]
    ax.plot(eps, acc, linestyle=style, linewidth=1.5, label=label)
ax.set_xlabel("Epoch")
ax.set_ylabel("Validation accuracy")
ax.set_ylim(0.0, 0.75)
ax.grid(True, linestyle="--", alpha=0.4)
ax.legend(loc="lower right", fontsize=9)
ax.set_title("Validation accuracy across models (grouped split)")
fig.tight_layout()
p = OUT_DIR / "curve_val_accuracy_comparison.png"
plt.savefig(p, dpi=150, bbox_inches="tight")
plt.show()
print("Wrote", p)
```
